# Download Image

This notebook downloads the flower image and saves it as `success.jpg`.

```python
import requests
from pathlib import Path

image_url = "https://bouqs.com/blog/wp-content/uploads/2022/03/shutterstock_260182148-min.jpg"
output_path = Path("success.jpg")

headers = {
    "User-Agent": "Mozilla/5.0"
}

response = requests.get(image_url, headers=headers, timeout=30)

output_path.write_bytes(response.content)

print(f"Downloaded image to: {output_path.resolve()}")
```
